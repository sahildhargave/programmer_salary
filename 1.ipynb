```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("survey_results_publi.csv")
```

```python
df.head()
```

```python
df=df[["Country", "EdLevel", "YearsCodePro", "Employment", "ConvertedComp"]]
df=df.rename({"ConvertedComp":"Salary"}, axis=1)
df.head()
```

```python
df=df[df["Salary"].notnull()]
df.head()
```

```python
#ðŸ˜Ž To Get information about the Database
#  DataFrame provides valuable information 
# about the DataFrame's structure and data types
df.info()
```

```python
#ðŸ˜ŽðŸ˜ŽðŸ˜ŽðŸ˜Ž
df=df.dropna()
df.isnull().sum()
```

```python
df=df[df["Employment"] == "Employed full-time"]
df=df.drop("Employment",axis=1)
df.info
```

```python
df["Country"].value_counts()
```

```python
def shorten_categories(categories, cutoff):
    categorical_map={}
    for i in range(len(categories)):
       if categories.values[i] >= cutoff:
           categorical_map[categories.index[i]] =categories.index[i]
    else:
        categorical_map[categories.index[i]] = 'Other'
    return categorical_map
```

```python
country_map = shorten_categories(df.Country.value_counts(),400)
df['Country'] = df['Country'].map(country_map)
df.Country.value_counts()
```

```python
fig, ax = plt.subplots(1,1, figsize=(12, 7))
df.boxplot('Salary','Country',ax=ax)
plt.suptitle('Salary (US$) v Country')
plt.title('')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
df= df[df["Salary"] <= 250000]
df= df[df["Salary"] >= 10000]
df= df[df["Country"] != 'Other']
```

```python
fig, ax = plt.subplots(1,1, figsize=(12, 7))
df.boxplot('Salary', 'Country', ax=ax)
plt.suptitle('Salary (US$) v Country')
plt.title('')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
df["YearsCodePro"].unique()
```

```python
def clean_experience(x):
    if x ==  'More than 50 years':
        return 50
    if x == 'Less than 1 year':
        return 0.5
    return float(x)

df['YearsCodePro'] = df['YearsCodePro'].apply(clean_experience)
```

```python
df["EdLevel"].unique()
```

```python
def clean_education(x):
    if 'Bachelorâ€™s degree' in x:
        return 'Bachelorâ€™s degree'
    if 'Masterâ€™s degree' in x:
        return 'Masterâ€™s degree'
    if 'Professional degree' in x or 'Other doctoral' in x:
        return 'Post grad'
    return 'Less than a Bachelors'

df['EdLevel'] = df['EdLevel'].apply(clean_education)
```

```python
df["EdLevel"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le_education = LabelEncoder()
df['EdLevel'] = le_education.fit_transform(df['EdLevel'])
df['EdLevel'].unique()
```

```python
le_country = LabelEncoder()
df['Country'] = le_country.fit_transform(df['Country'])
df["Country"].unique()
```

```python
x= df.drop("Salary",axis=1)
y= df["Salary"]
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(x, y.values)
```

```python
y_pred = linear_reg.predict(x)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
error =np.sqrt(mean_squared_error(y, y_pred)) # y target value , y_pred value
```

```python
error
```

```python
from sklearn.tree import DecisionTreeRegressor
dec_tree_reg = DecisionTreeRegressor(random_state=0)
dec_tree_reg.fit(x, y.values)
```

```python
y_pred = dec_tree_reg.predict(x)
```

```python
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```

```python
from sklearn.ensemble import RandomForestRegressor
random_forest_reg = RandomForestRegressor(random_state=0)
random_forest_reg.fit(x, y.values)
```

```python
y_pred = random_forest_reg.predict(x)
```

```python
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```

```python
from sklearn.model_selection import GridSearchCV

max_depth = [None, 2,4,6,8,10,12]
parameters = {"max_depth": max_depth}

regressor = DecisionTreeRegressor(random_state=0)
gs = GridSearchCV(regressor, parameters, scoring='neg_mean_squared_error')
gs.fit(x, y.values)
#the negative mean squared error is commonly used as a loss function to be minimized
```

```python
regressor = gs.best_estimator_

regressor.fit(x, y.values)
y_pred = regressor.predict(x)
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```
